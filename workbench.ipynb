```python
from cdptools.databases.cloud_firestore_database import CloudFirestoreDatabase
from cdptools.file_stores.gcs_file_store import GCSFileStore
from cdptools.utils import research_utils
import json
import pandas as pd
import tempfile

db = CloudFirestoreDatabase("stg-cdp-seattle")
fs = GCSFileStore("stg-cdp-seattle.appspot.com")
db, fs
```

```python
transcript_manifest = research_utils.get_most_recent_transcript_manifest(db)
transcript_manifest.head()
```

```python
with tempfile.TemporaryDirectory() as tmpdir:
    for i, row in transcript_manifest.iterrows():
        transcript = fs.download_file(row["uri"], tmpdir)
        with open(transcript, "r") as read_in:
            transcript = json.load(read_in)
        
        raw = [s["sentence"] for s in transcript]
        raw = " ".join(raw)
        print(raw[:100])
        break
```
